## CSS를 이용하여 tag 찾기
- select, select_one 함수 사용

```python
# bs4 = BeautifulSoup 4 버전
from bs4 import BeautifulSoup
```

### CSS selector 사용법
- 태그명 찾기 tag
- 자손 태그 찾기 - 자손 관계(tag tag)
- 자식 태그 찾기 - 다이렉트 자식 관계(tag > tag) : 최초의 '바로 하위' 단계를 의미한다.
- 아이디 찾기 #id
- 클래스 찾기 .class
- 속성값 찾기 [name='test']
    - 속성값 prefix 찾기 [name ^='test'] : 부분 문자열(틸드)
    - 속성값 suffix 찾기 [name $='test']
    - 속성값 substring 찾기 [name *='test']
- n번째 자식 tag 찾기 : nth-child(n)

```python
# 태그명 찾기 tag
url = 'https://news.v.daum.net/v/20190728165812603'

response = requests.get(url)
soup = BeautifulSoup(response.text)

soup.select('h3')
```

```python
# id가 harmonyContainer이고, 그 안의 모든 p를 찾아라(자손은 띄어쓰기, 자식은 부등호 '>')
soup.select('#harmonyContainer p') 
```

```python
# 클래스 명은 '.'을 찍는다
soup.select('h3.tit_view')
```

```python
# 클래스는 이렇게도 찾는다.
soup.select('h3[class="tit_view"]')
```

```python
# 'tx'로 시작하는 클래스를 찾아라.
soup.select('h3[class^="tx"]')
```

```python
# 'view'로 끝나는 클래스를 찾아라.
soup.select('h3[class$="view"]')
```

```python
# '_'를 사용하는 모든 클래스를 찾아라.
soup.select('h3[class*="_"]')
```

```python
# 특정 순서에 있는 요소 찾기1
soup.select('span.txt_info')[0]
```

```python
# 특정 순서에 있는 요소 찾기2
soup.select('span.txt_info::nth-child(1)')
```

### 뉴스 제목 크롤링

```python
# 함수 형태로 개발!
def get_daum_news_title(news_id):
    url = 'https://news.v.daum.net/v/{}'.format(news_id)
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    title_tag = soup.select_one('h3.tit_view')
    
    if title_tag:
        return title_tag.get_text()
    return ""
```

```python
get_daum_news_title('20190728165812603')
```

### 뉴스 본문 크롤링

```python
def get_daum_news_content(news_id):
    url = 'https://news.v.daum.net/v/{}'.format(news_id)
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    
    content = ''
    for p in soup.select('div#harmonyContainer p'):
        content += p.get_text()
    return content
```
